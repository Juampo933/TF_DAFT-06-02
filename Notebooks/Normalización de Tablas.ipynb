Creación de tabla Detalle_de_Compras:

```python
PurchasesFINAL12312016.head(3)
```

```python
Detalle_de_Compras =  PurchasesFINAL12312016.drop(['VendorNumber', 'VendorName', 'Classification'], axis=1)
Detalle_de_Compras.head(3)
```

```python
# Calcular la moda (valor más frecuente) de la columna 'Size'
moda_size = Detalle_de_Compras['Size'].mode()[0]  # mode()[0] obtiene el valor más frecuente

# Rellenar los valores NaN con la moda
Detalle_de_Compras['Size'].fillna(moda_size, inplace=True)

# Verificar si quedan NaN
Detalle_de_Compras['Size'].isnull().sum()
```

Codificamos las variables Description y Size que están como texto, a un valor numérico, para combinarlas con Brand y tener un código único de producto:

```python
from sklearn.preprocessing import LabelEncoder

# Crear un objeto de LabelEncoder
label_encoder = LabelEncoder()
```

```python
# Aplicar el encoder a la columna 'Description'
Detalle_de_Compras['Description_encoded'] = label_encoder.fit_transform(Detalle_de_Compras['Description'])

# Mostrar las primeras filas para verificar la codificación
Detalle_de_Compras[['Description', 'Description_encoded']].head()
```

```python
# Aplicar el encoder a la columna 'Size'
Detalle_de_Compras['Size_encoded'] = label_encoder.fit_transform(Detalle_de_Compras['Size'])

# Mostrar las primeras filas para verificar la codificación
Detalle_de_Compras[['Size', 'Size_encoded']].head()
```

```python
Detalle_de_Compras['ProductoID'] = (Detalle_de_Compras['Brand'].astype(str) + 
                       Detalle_de_Compras['Description_encoded'].astype(str) + 
                       Detalle_de_Compras['Size_encoded'].astype(str)).astype(int)

Detalle_de_Compras.head()
```

```python
Detalle_de_Compras.ProductoID.nunique()
```

Creación de Tabla Producto:

```python
# Eliminar duplicados basados en el ID
Productos = Detalle_de_Compras.drop_duplicates(subset='ProductoID')[["ProductoID","Brand","Description","Description_encoded","Size","Size_encoded"]]

# Ordenando por la columna 'VendorNumber' en orden ascendente
Productos = Productos.sort_values('ProductoID').reset_index(drop=True)

# Diccionario que para cambiar nombres de las columnas 
column_renames = {
    'Brand': 'Marca',
    'Description': 'Descripción',
    'Description_encoded': 'DescripciónID',
    'Size': 'Tamaño',    
    'Size_encoded': 'TamañoID'
}
# Renombrar las columnas
Productos.rename(columns=column_renames, inplace=True)

Productos.head(2)
```
